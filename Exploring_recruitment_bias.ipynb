```python
#importing libraries
import pandas as pd
import random
```

```python
number_candidates = 50
```

```python
candidates=[]
```

```python
#Generating random candidate data using the random library
for i in range(number_candidates): 
  gender= random.choice(["M", "F"])
  age=  random.randint(18,50)
  years_of_experience= random.randint(0,15)
  skill_score=random.random()
  work_ethic=random.random()
  attitude_score=random.random()
  ethnicity= random.choice(["Black", "White", "Asian", "Arab", "Other"])
  disability=random.choice([0,1])
  legal_history=random.choice([0,1])
  country_of_origin=random.randint(0,217)
  time_spent_viewing=random.randint(3,120)
  hired =random.choice([0,1])
  candidates.append([gender, age, years_of_experience, skill_score, work_ethic, attitude_score,ethnicity, disability, legal_history, country_of_origin, time_spent_viewing, hired])
```

```python
candidates
```

```python
#Converting the candidates matrix to a dataframe
candidates_df = pd.DataFrame(candidates, columns=['gender','age', 'years_of_experience','skill_score', 'work_ethic','attitude_score', 'ethnicity', 'disability', 'legal_history','country_of_origin', 'time_spent_viewing', 'hired' ]) 
```

```python
candidates_df
```

```python
#Encoding categorical variables
gender_mapper = {"M":1, "F":0}
ethnicity_mapper = {"Black":0, "White":1, "Arab":2, "Asian":3, "Other":4}

candidates_df["gender"] = candidates_df["gender"].replace(gender_mapper)
candidates_df["ethnicity"] = candidates_df["ethnicity"].replace(ethnicity_mapper)
```

```python
candidates_df
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#A simple model for prototyping
lr = LogisticRegression(max_iter=1000)
```

```python
#Splitting the candidates data to input and target features
columns = ['gender','age', 'years_of_experience','skill_score', 'work_ethic','attitude_score', 'ethnicity', 'disability', 'legal_history','country_of_origin', 'time_spent_viewing']
X = candidates_df[columns]
y = candidates_df['hired']
```

```python
#Fitting the model
lr.fit(X,y)
```

```python
#Determining feature importance by printing coefficients
lr.coef_
```
